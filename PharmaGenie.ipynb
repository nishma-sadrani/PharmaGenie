```python
import numpy as np
import pandas as pd
```

```python
medicines = pd.read_csv(r"/content/medicine.csv")
```

```python
medicines.head()
```

```python
medicines.isnull().sum()
```

```python
medicines.dropna(inplace=True)
```

```python
medicines.duplicated().sum()
```

```python
medicines['Description']
```

```python
medicines['Description'].apply(lambda x:x.split())
```

```python
medicines['Reason'] = medicines['Reason'].apply(lambda x:x.split())
medicines['Description'] = medicines['Description'].apply(lambda x:x.split())
```

```python
medicines
```

```python
medicines['Description'] = medicines['Description'].apply(lambda x:[i.replace(" ","") for i in x])
medicines['Reason'] = medicines['Reason'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
medicines['tags'] = medicines['Description'] + medicines['Reason']
```

```python
medicines_new = medicines[['index','Drug_Name','tags']]
```

```python
medicines_new
```

```python
medicines_new['tags'] = medicines_new['tags'].apply(lambda x:" ".join(x))
```

```python
medicines_new
```

```python
medicines_new['tags'] = medicines_new['tags'].apply(lambda x:x.lower())
```

```python
medicines_new
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(stop_words='english',max_features=5500)
```

```python
def stem(text):
  y = []

  for i in text.split():
    y.append(ps.stem(i))

  return " ".join(y)
```

```python
medicines_new['tags'] = medicines_new['tags'].apply(stem)
```

```python
medicines_new
```

```python
cv.fit_transform(medicines_new['tags']).toarray().shape
```

```python
vectors = cv.fit_transform(medicines_new['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names_out
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors)
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[20]
```

```python
def recommend(medicine):
  medicine_index = medicines_new[medicines_new['Drug_Name'] == medicine].index[0]
  distances = similarity[medicine_index]
  medicines_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

  recommended_medicines = []
  for i in medicines_list:
    recommended_medicines.append(medicines_new.iloc[i[0]].Drug_Name)

  return recommended_medicines
```

```python
recommend("Wokadine M Onit 10gm")
```

```python
# pip install gradio
```

```python
import gradio as gr

iface = gr.Interface(
    fn=recommend,
    inputs=gr.Dropdown(choices=medicines_new.Drug_Name.tolist(), label="Enter a medicine name"),
    outputs=gr.List(label="Recommended Medicines"),
    title="Medicine Recommendation System",
    description="Start typing a medicine name to see suggestions and get recommendations"
)

iface.launch()
```
